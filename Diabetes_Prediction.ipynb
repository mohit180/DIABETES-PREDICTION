```python
# importing all required dependendcies
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler # to standardized the data
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
# data collection and analysis
# PIMA Diabetes Dataset
# loading the diabetes dataset to a pandas DataFrame
diabetes_dataset = pd.read_csv('/content/diabetes.csv')
```

```python
# saparating data and labels
X= diabetes_dataset.drop(columns = 'Outcome',axis=1)
Y = diabetes_dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

DATA STANDARDIZATION - TO DEAL WITH INCONSISTENT DATA

```python
scaler= StandardScaler()
```

```python
scaler.fit(X)
```

```python
standardized_data=scaler.transform(X)
```

```python
print(standardized_data)
```

NOW ALL VALUES ARE IN SAME RANGE I.E BETWEEN O AND 1 THIS WILL HELP OUR MODEL TO MAKE BETTER PREDICTION

```python
X=standardized_data
Y= diabetes_dataset['Outcome']
```

```python
print(X)
print(Y)
```

TRAIN TEST SPLITTING

```python
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

TRAINING THE MODEL

```python
classifier = svm.SVC(kernel='linear')
```

```python
classifier.fit(X_train,Y_train)
```

MODEL EVALUATION :

```python
# accuracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('accuracy score of training data :' , training_data_accuracy)
```

```python
# accuracy score on test data
X_test_prediction = classifier.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("accuracy score for test data :", test_data_accuracy)
```

MAKING A PREDICTIVE SYSTEM TO PREDICT WHETHER A PATIENT IS DIABETIC OR NOT I.E. LABEL BASED ON FEATURES GIVEN

```python
input_data=(1,89,66,23,94,28.1,0.167,21) # it should predict zero ie non diabetic
# changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# we need to standardize the input_data just like training data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if prediction[0]==0 : # we need to mention index because it is list
  print(" the person is not diabetic")
else :
  print(" the person is diabetic")



```

OUR PREDICTIVE SYSTEM IS WORKING FINE BECAUSE MODEL IS PREDICTING CORRECTLY FOR THE GIVEN FEATURES
